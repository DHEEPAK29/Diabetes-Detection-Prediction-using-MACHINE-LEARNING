```python
X=df1[['Number_pregnant','Glucose_concentration','Blood_pressure','Triceps','Insulin','BMI','Pedigree','Age']].values
X
```

```python
y=df1['Class'].values
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
DT=DecisionTreeClassifier()
DT
```

```python
DT.fit(X_train, y_train)
```

```python
prediction=DT.predict(X_test)
prediction
```

```python
from sklearn.metrics import confusion_matrix
```

```python
CMDT=confusion_matrix(y_test,prediction)
CMDT
```

```python
CMDTexpand = pd.crosstab(y_test, prediction, rownames=['Actual'], colnames=['Predicted'], margins=True)
CMDTexpand
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,prediction)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,prediction))
```

```python

RF=RandomForestClassifier()
```

```python
RF.fit(X_train, y_train)
```

```python
PredictionRF=RF.predict(X_test)
PredictionRF
```

```python
from sklearn.metrics import confusion_matrix
```

```python
CMRF=confusion_matrix(y_test,PredictionRF)
CMRF
```

```python
CMRFexpand = pd.crosstab(y_test, PredictionRF, rownames=['Actual'], colnames=['Predicted'], margins=True)
CMRFexpand
```

```python
accuracy_score(y_test,PredictionRF)
```

```python

from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,PredictionRF))
```
